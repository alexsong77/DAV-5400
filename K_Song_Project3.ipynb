```python
import pandas as pd
import numpy as np 
# read csv from github 'Sunqi'(enrollemnt)
enrollment = pd.read_csv('https://raw.githubusercontent.com/alexsong77/DAV-5400/master/Sunqi.csv')
enrollment
```

**1.Purpose: Analyze the enrollment on different ethnicity by the level, or compare different institutions’ enrollment by ethnicity.**

```python
# drop the none value 
enrollment.drop(columns=['Unnamed: 9'],inplace=True)
enrollment.dropna(how='all',inplace=True)

enrollment
```

```python
# melt the wide format in to long format 
enrollment1 = enrollment.melt(id_vars=['Unique IDof the institution', 'Level of student'], value_vars=['American Indian or Alaska Native','Asian','Black or African American','Hispanic or Latino','Native Hawaiian or Other Pacific Islander','White','Two or more races','Race/ethnicity unknown','Nonresident alien'])
enrollment1.head()
```

```python
#renamed the column name
enrollment1.columns = ['ID','Level of student','Ethnicity','Number of students']
enrollment1.head
```

```python
#group each dimensions by main categorizes 
enrollment2 = enrollment1.groupby([enrollment1['Level of student'],enrollment1['Ethnicity']],as_index=False)['Number of students'].sum()
enrollment2
```

```python
#group each dimensions by main categorizes "Level of student", "Ethnicity", and "Nummber of students"
enrollment2 = enrollment1.groupby([enrollment1['Level of student'],enrollment1['Ethnicity']])['Number of students'].sum().to_frame()
enrollment2
```

```python
# find out which level of student and ethnicity has the highest number of enrollment students 
print(enrollment2.idxmax())
print(enrollment2['Number of students'].max())
```

```python
# find out which level of student and ethnicity has the lowest number of enrollment students 
print(enrollment2.idxmin())
print(enrollment2['Number of students'].min())
```

**Conclusion:for each level of student<br> "Undergraudete - White" has the highest number of enrollment students: 12773 <br>
                "Graduate - Native Hawaiian or Other Pacific Islander" has the lowest number of enrollment students:16**
            

```python
#group each dimensions by main categorizes "ID", "Ethnicity", and "Nummber of students"
enrollment3 = enrollment1['Number of students'].groupby([enrollment1['ID'],enrollment1['Ethnicity']]).sum().to_frame()
enrollment3
```

```python
# find out which ID and ethnicity has the highest number of enrollment students 
print(enrollment3.idxmax())
print(enrollment3['Number of students'].max())
```

```python
# find out which ID and ethnicity has the lowest number of enrollment students 
print(enrollment3.idxmin())
print(enrollment3['Number of students'].min())
```

**Conclusion:for each ID of institution<br> "ID100663 - White" has the highest number of enrollment students: 13952 <br>
                "ID100690 - Nonresident alien" has the lowest number of enrollment students:0**

```python
# read csv from github “Natan” (grades)
grades = pd.read_csv('https://raw.githubusercontent.com/alexsong77/DAV-5400/master/Natan%20-%20Test%20Grades.csv')
grades
```

**3.Purpose:<br> 1. Which student had the highest average test grades?<br>2. Which student had the greatest range of test grades?<br>3. Which test had the highest grades?**

 

```python
# drop the none value 
grades.dropna(how='all',inplace=True)
grades
```

```python
# make up the non value by filling the previous columns
grades['Test 1'].fillna(method='ffill', inplace=True)
grades['Test 3'].fillna(method='ffill', inplace=True)
grades
```

```python
# melt the wide format in to long format 
melted1 = pd.melt(grades, ['Name'])
melted1.head()
```

```python
#renamed the columns
melted1.columns = ['Name','Test','Value']
melted1.head()
```

```python
#calculated the average test grades and groupby each categorizes by name, and then sorted the value 
name = melted1.groupby('Name').mean()
name.sort_values(by='Value',ascending=False)
```

**1.Jim Jackson had the highest average test grades**

```python
# groupby and calculated each name ranges 
range1 = melted1.groupby('Name').max()['Value']-melted1.groupby('Name').min()['Value']
range1
```

```python
# sorted the ranges of test grades
range1.sort_values(ascending=False)
```
